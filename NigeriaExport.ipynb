```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
```

Reading in the dataset

```python
df = pd.read_csv("./nigeria_agricultural_exports.csv", skipinitialspace= True)
df.head()
```

```python
df.columns
```

Renaming column names

```python
df = df.rename(columns={"Product Name":"Product_Name", "Export Country": "Export_Country", "Units Sold": "Units_Sold", 
                        "Profit per unit": "Profit_per_unit", "Export Value": "Export_Value", 
                        "Destination Port": "Destination_Port", "Transportation Mode": "Transportation_Mode"})
df.head()
```

Data Cleaning

Checking for null values

```python
df.isna().sum()
```

Checking for duplicate values

```python
df.duplicated().sum()
```

Calculating the revenue

```python
df["Revenue"] = df["Units_Sold"] * df["unit_price"]
df.head()
```

```python
df["Average_Revenue"] = df["Revenue"].mean()
df.head()
```

```python
df.dtypes
```

changing datatypes

Top best selling Products

```python
# Set the styling
plt.style.use("fivethirtyeight")

# getting the two columns for the analysis into a new dataset
df1 = df[["Product_Name", "Revenue"]]
df1.sort_values(by= "Revenue", ascending= False, inplace= True)

# Get the top 5 revenue earners
top_df = df1.nlargest(4, "Revenue")  # Use nlargest for top N

# Separate product names and revenue into lists
product_names = top_df["Product_Name"].tolist()
revenue_values = top_df["Revenue"].tolist()

# Plot the bar chart
plt.figure(figsize=(12, 6)) 
plt.bar(product_names, revenue_values, color="#008fd5")
plt.legend()
plt.title("Top Best Selling Products")
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.xticks(wrap=True)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
month_map = {"January": 1, "February": 2, "March": 3, "April": 4, "May": 5, "June": 6,
             "July": 7, "August": 8, "September": 9, "October": 10, "November": 11, "December": 12}

df["month_num"] = df["month"].map(month_map)  # Create a new column with numerical month values
```

How do sales vary over time(monthly, quarterly, annually)

```python
# Define functions for plotting yearly, quarterly, and monthly variations
def plot_yearly_sales(df):
  """Plots total sales per year"""
  yearly_sales = df.groupby('year')['Revenue'].sum()
  plt.figure(figsize=(10, 6))
  plt.plot(yearly_sales.index, yearly_sales.values)  # Extract single values from Series
  plt.xlabel("Year")
  plt.ylabel("Total Sales")
  plt.title("Yearly Sales")
  plt.grid(True)
  plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
  plt.tight_layout()

def plot_quarterly_sales(df):
  """Plots total sales per quarter"""
  quarterly_sales = df.groupby(df['quarter'])['Revenue'].sum()
  plt.figure(figsize=(10, 6))
  plt.plot(quarterly_sales.index, quarterly_sales.values)  # Extract single values from Series
  plt.xlabel("Quarter")
  plt.ylabel("Total Sales")
  plt.title("Quarterly Sales")
  plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
  plt.grid(True)
  plt.tight_layout()

def plot_monthly_sales(df):
  """Plots total sales per month, ensuring January starts the axis"""
  monthly_sales = df.groupby('month_num')['Revenue'].sum()
  # Sort monthly sales by month_num (numerical)
  monthly_sales = monthly_sales.sort_index()
  plt.figure(figsize=(12, 6))
  plt.plot(monthly_sales.index, monthly_sales.values)  # Extract single values from Series
  plt.xlabel("Month")
  plt.xticks([i for i in range(1, 13)], monthly_sales.index)  # Use numerical month values for x-axis ticks
  plt.ylabel("Total Sales")
  plt.title("Monthly Sales")
  plt.grid(True)
  plt.tight_layout()

# Call the plotting functions
plot_yearly_sales(df.copy())  # Avoid modifying original DataFrame
plot_quarterly_sales(df.copy())
plot_monthly_sales(df.copy())

# Display the plots (optional)
plt.show()
```

Cost Analysis

Cost of goods csold as a precentage of revenue

```python
df['cost_of_goods_sold'] = df['Units_Sold'] * df['Profit_per_unit']
df['Cogs_Percentage'] = (df['cost_of_goods_sold'] / df['Revenue']) * 100
avg_cogs = df['Cogs_Percentage'].mean()
print("The Average COGS Percentage:", avg_cogs)
```

Cost of goods across different products

```python
cogs_group = df.groupby('Product_Name')['cost_of_goods_sold'].sum()
cogs = cogs_group.sort_values(ascending=False)
plt.figure(figsize=(12,6))
cogs.plot(kind='bar')
plt.title('Total Cost of Goods Sold (COGS) by Product')
plt.xlabel('Product')
plt.ylabel('Total COGS')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
